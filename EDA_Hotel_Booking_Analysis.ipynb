#**Project Summary-**


In the hotel industry, understanding customer behavior and optimizing booking strategies is crucial for maximizing revenue and ensuring guest satisfaction. This project addresses several key business challenges:

2-Optimal Length of Stay: Determining the ideal duration for a guest's stay that offers the most cost-effective pricing is essential. This insight can help hotels optimize room utilization and revenue.

4-Revenue Management: Effective revenue management is a critical challenge in the hotel industry. By analyzing data, hotels can better forecast demand, optimize pricing strategies, and improve overall revenue.

5-Customer Experience: Providing a personalized and seamless guest experience is a key competitive advantage. Understanding customer booking behavior and preferences can enhance the guest experience and drive repeat business.

# ***3. Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
hotel_df = hotel_df.drop_duplicates()
hotel_df
```

```python
#Replacing Null values
null_columns=['company','agent','children']
for columns in null_columns:
  hotel_df[columns].fillna(0,inplace=True)

#Replacing the NA values of country column with others
hotel_df['country'].fillna('others',inplace=True)
```

```python
# Null values have been removed or filled
hotel_df.isnull().sum().reset_index().rename(columns={'index':'Columns',0:'Null values'})
```

### **Adding New Columns to the Dataset**

```python
# Adding total number of people stayed in the hotels
hotel_df['total_people'] = hotel_df['adults'] + hotel_df['babies'] + hotel_df['children']

# Adding Total staying days in the hotels
hotel_df['total_stay'] = hotel_df['stays_in_weekend_nights'] + hotel_df['stays_in_week_nights']
```

```python
#Changing the data type of total people column from float to int
hotel_df['total_people'] = hotel_df['total_people'].astype(int)
```

```python
# Dataset view after adding the columns
hotel_df.head()
```

### There are some rows in total number of adults, children or babies equal to zero this means no booking were made. So we can remove such rows.

```python
# shape of columns which have no bookings
hotel_df[hotel_df['adults']+hotel_df['babies']+hotel_df['children'] == 0].shape
```

```python
# Columns are dropped here using drop function
hotel_df.drop(hotel_df[hotel_df['adults']+hotel_df['babies']+hotel_df['children'] == 0].index, inplace = True)
```

```python
# All null values have been removed
hotel_df.isna().sum()
```

### Chart - 8

### **8.In which month most of the bookings happened?**

```python
# Chart - 8 visualization code
bookings_by_months=hotel_df.groupby(['arrival_date_month'])['hotel'].count().reset_index().rename(columns={'hotel':"Counts of booking"}).sort_values('Counts of booking' , ascending = False)
bookings_by_months
```

```python
bookings_by_months=hotel_df.groupby(['arrival_date_month'])['hotel'].count().reset_index().rename(columns={'hotel':"Counts of booking"})
sequence_of_months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
bookings_by_months['arrival_date_month']=pd.Categorical(bookings_by_months['arrival_date_month'],categories=sequence_of_months,ordered=True)
bookings_by_months=bookings_by_months.sort_values('arrival_date_month')
bookings_by_months
```

```python
# Visualization of month in which most booking happened using Barplot.
plt.figure(figsize=(14,7))
sns.barplot(data=bookings_by_months, x="arrival_date_month", y="Counts of booking")
plt.title("Number of Bookings in Months", fontsize = 20)
plt.xlabel('Month', fontsize = 15)
plt.ylabel('Number of Bookings', fontsize = 15)
```
