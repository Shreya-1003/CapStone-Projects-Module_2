# ***3. Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
hotel_df = hotel_df.drop_duplicates()
hotel_df
```

```python
#Replacing Null values
null_columns=['company','agent','children']
for columns in null_columns:
  hotel_df[columns].fillna(0,inplace=True)

#Replacing the NA values of country column with others
hotel_df['country'].fillna('others',inplace=True)
```

```python
# Null values have been removed or filled
hotel_df.isnull().sum().reset_index().rename(columns={'index':'Columns',0:'Null values'})
```

### **Adding New Columns to the Dataset**

```python
# Adding total number of people stayed in the hotels
hotel_df['total_people'] = hotel_df['adults'] + hotel_df['babies'] + hotel_df['children']

# Adding Total staying days in the hotels
hotel_df['total_stay'] = hotel_df['stays_in_weekend_nights'] + hotel_df['stays_in_week_nights']
```

```python
#Changing the data type of total people column from float to int
hotel_df['total_people'] = hotel_df['total_people'].astype(int)
```

```python
# Dataset view after adding the columns
hotel_df.head()
```

### There are some rows in total number of adults, children or babies equal to zero this means no booking were made. So we can remove such rows.

```python
# shape of columns which have no bookings
hotel_df[hotel_df['adults']+hotel_df['babies']+hotel_df['children'] == 0].shape
```

```python
# Columns are dropped here using drop function
hotel_df.drop(hotel_df[hotel_df['adults']+hotel_df['babies']+hotel_df['children'] == 0].index, inplace = True)
```

```python
# All null values have been removed
hotel_df.isna().sum()
```

### Chart - 2

### 2. Which agent made the most bookings?


```python
#chart-2 visualization code
highest_bookings = hotel_df.groupby(['agent'])['agent'].agg({'count'}).reset_index().rename(columns={'count': "Most_Bookings"}).sort_values(by='Most_Bookings',ascending=False)
highest_bookings.drop(highest_bookings[highest_bookings['agent']==0].index,inplace=True)

# taking top 10 bookings made by agent
top_ten_highest_bookings=highest_bookings[:10]

top_ten_highest_bookings
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=top_ten_highest_bookings['agent'],y=top_ten_highest_bookings['Most_Bookings'],order=top_ten_highest_bookings['agent'])
plt.title('Most bookings made by the agent', fontsize=15)
plt.ylabel('Number of bookings', fontsize=15)
plt.xlabel('Agent number', fontsize=15)
plt.show()
```

**I choose barplot here because it provides a visual representation of the data, making it easy to compare.**
